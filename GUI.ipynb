```python
import tkinter as tk
from tkinter import messagebox
import requests

# Function to fetch lyrics from Lyrics.ovh API
def get_lyrics():
    song = song_entry.get()
    artist = artist_entry.get()

    if song and artist:
        url = f"https://api.lyrics.ovh/v1/{artist}/{song}"
        response = requests.get(url)

        if response.status_code == 200:
            lyrics = response.json().get('lyrics', 'Lyrics not found!')
            lyrics_text.delete(1.0, tk.END)  # Clear previous text
            lyrics_text.insert(tk.END, lyrics)
        else:
            messagebox.showerror("Error", "Lyrics not found! Check the song and artist name.")
    else:
        messagebox.showwarning("Input Error", "Please enter both song and artist name.")

# GUI setup using Tkinter
root = tk.Tk()
root.title("Lyrics Extractor")
root.geometry("500x400")

# Labels and Entries for song and artist input
song_label = tk.Label(root, text="Song Title:", font=("Helvetica", 12))
song_label.pack(pady=10)

song_entry = tk.Entry(root, width=40)
song_entry.pack(pady=5)

artist_label = tk.Label(root, text="Artist Name:", font=("Helvetica", 12))
artist_label.pack(pady=10)

artist_entry = tk.Entry(root, width=40)
artist_entry.pack(pady=5)

# Button to fetch lyrics
fetch_button = tk.Button(root, text="Fetch Lyrics", command=get_lyrics, font=("Helvetica", 12), bg="lightblue")
fetch_button.pack(pady=20)

# Text box to display the lyrics
lyrics_text = tk.Text(root, height=10, width=60, wrap=tk.WORD)
lyrics_text.pack(pady=10)

# Start the GUI loop
root.mainloop()
```
